# Narcolepsy Model

This is example code of how to classify narcoleptic patients using electronic health record data.  
EHR data needed:
- Clinical visit notes
- ICD codes
- Medications

```python
import polars as pl
import pandas as pd
from narcolepsy_model import NarcolepsyModel
from pathlib import Path

model = NarcolepsyModel(
    model_type='nt12_vs_not' # or 'nt1_vs_not', 'nt2_vs_not'
)

data_path = Path('data') # path to data directory

data = {
    'note' : pl.read_parquet(data_path / 'note.parquet').rename({'bdsp_patient_id':'id'}).cast({'id': pl.Utf8}),
    'icd' : pl.read_parquet(data_path / 'icd.parquet').rename({'bdsp_patient_id':'id'}).cast({'id': pl.Utf8}),
    'med' : pl.read_parquet(data_path / 'med.parquet').rename({'bdsp_patient_id':'id'}).cast({'id': pl.Utf8})
}


# can use pandas dataframes as well
# output will be polars dataframes, but easy to convert back to pandas if needed

# data = {
#     'note' : pd.read_parquet(data_path / 'note.parquet').rename(columns={'bdsp_patient_id':'id'}).astype({'id': str}),
#     'icd' : pd.read_parquet(data_path / 'icd.parquet').rename(columns={'bdsp_patient_id':'id'}).astype({'id': str}),
#     'med' : pd.read_parquet(data_path / 'med.parquet').rename(columns={'bdsp_patient_id':'id'}).astype({'id': str})
# }
```

```python
# this is the schema used by by the feature extraction pipeline
# ensure your data matches this schema
model.config['schema']
```

## features and predictions

```python
feat, pred = model.run(
    data,
    show_progress=True,
    return_features=True
)
```

## model training

```python
from model_comp import leave_one_source_out_validation, define_models_config, plot_leave_one_out_curves, train_final_production_model
import ast
```

```python
# feature matrix and labels for evaluation
X = data['note'].select('annot', 'cohort').hstack(feat).drop('id', 'date').with_columns(
    pl.col('annot').replace({
        # annotation scheme is
        # 1: NT1
        # 2: NT2 / IH
        # 3: Narcolepsy, unsure
        # 4: Not narcolepsy

        # for this example, we will do NT12 (any type) vs Not training
        1:1,
        2:1,
        3:1,
        4:0
    })
)
X.head()
```

```python
models_config = define_models_config()
models_config.keys()

# you can drop any models you don't want to test
# models_config.pop('GradientBoosting')
```

```python
# grid search, leave one cohort out validation
results, curves_data = leave_one_source_out_validation(
    X,
    source_col="cohort",
    target_col="annot",
    models_config=models_config,
    scoring_metric="average_precision",
    output_dir="results",
    random_state=42,
    # save_fold_models=True # uncomment to save models for each fold
)
```

## results

```python
results.write_csv('results/results.csv')
with pl.Config(tbl_rows=-1):
    display(results)
```

```python
plot_leave_one_out_curves(
    curves_data,
    results,
    output_dir='results'
)
```

### train final model (using all of the data)

```python
# the metrics above are from cross-validation
# can use best hyperparameters from model of choice to train final model

ast.literal_eval(results.filter(pl.col('model') == 'RandomForest').sort('f1', descending=True)[0, 'best_params'])
```

```python
final_model, model_info = train_final_production_model(
    X,
    source_col="cohort",
    target_col="annot",
    best_model_config=(
        'RandomForest',
        models_config['RandomForest'][0],
        ast.literal_eval(results.filter(pl.col('model') == 'RandomForest').sort('f1', descending=True)[0, 'best_params'])
    ),
    output_dir="results/final_model",
    random_state=42
)
```
